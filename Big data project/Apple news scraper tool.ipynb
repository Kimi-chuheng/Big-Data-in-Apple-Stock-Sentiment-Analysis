```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup
import csv
import time

# 设置 ChromeDriver
options = webdriver.ChromeOptions()
options.add_argument("--headless")  # 运行在无头模式（不显示浏览器）
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)

# 打开目标页面
url = "https://www.apple.com/newsroom/archive/"
driver.get(url)

# 保存所有新闻数据
all_news_data = []

def get_news_items(page_source):
    """提取当前页面中的新闻项并保存到 all_news_data 列表"""
    # 使用 BeautifulSoup 解析 HTML
    soup = BeautifulSoup(page_source, 'lxml')

    # 查找新闻项
    news_items = soup.find_all('div', class_='item__content')

    # 提取新闻标题和时间
    for item in news_items:
        # 提取标题
        title_element = item.find('h3')
        title = title_element.text if title_element else 'No title'

        # 提取时间
        publishing_element = item.find('p', class_='item__date')
        time_info = publishing_element.text.strip() if publishing_element else 'No time info'

        # 将标题和时间保存到列表中
        all_news_data.append([title, time_info])

# 翻页循环，只抓取两页
page_count = 0
while page_count < 228:
    # 获取页面的 HTML
    html = driver.page_source

    # 获取当前页的新闻并保存
    get_news_items(html)
    
    # 输出已完成当前页
    print(f"Page {page_count + 1} completed.")
    
    # 尝试找到“下一页”按钮
    try:
        # 使用 aria-label 或者 class 定位“下一页”按钮
        next_button = driver.find_element("xpath", "//a[@aria-label='next News articles']")  # 或者使用完整类名

        if next_button.is_displayed():
            next_button.click()  # 点击下一页
            page_count += 1  # 计数
            time.sleep(3)  # 等待页面加载
        else:
            print("no more page")
            break
    except Exception as e:
        print(f"no button to next page,warning: {e}")
        break

# 关闭浏览器
driver.quit()

# 保存所有新闻数据到 CSV 文件
with open('apple_news.csv', 'w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(['Title', 'Time'])  # 写入表头
    writer.writerows(all_news_data)  # 写入数据

print("Data has been saved to apple_news.csv")
```

```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup
import csv
import time
import pyperclip
# 设置 ChromeDriver
options = webdriver.ChromeOptions()
options.add_argument("--headless")  # 运行在无头模式（不显示浏览器）
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)

# 打开目标页面
url = "https://www.apple.com/newsroom/archive/"
driver.get(url)

# 保存所有新闻数据
all_news_data = []

def get_news_items(page_source):
    """提取当前页面中的新闻项并保存到 all_news_data 列表"""
    # 使用 BeautifulSoup 解析 HTML
    soup = BeautifulSoup(page_source, 'lxml')

    # 查找新闻项
    news_items = soup.find_all('a', class_='result__item row-anchor')
    # 提取新闻标题和时间
    for item in news_items:

        title_element = item.find('h3')
        title = title_element.text if title_element else 'No title'

        # 提取时间
        publishing_element = item.find('p', class_='item__date')
        time_info = publishing_element.text.strip() if publishing_element else 'No time info'
        link = item['href'] if item else 'No link'
        news_content=get_news_content(link)

        # 返回到新闻列表页面
        all_news_data.append([title, time_info,news_content])
        
def get_news_content(link):
    """访问新闻链接并提取新闻内容"""
    # 使用完整链接访问新闻
    full_link = f"https://www.apple.com{link}"
    driver.get(full_link)
    time.sleep(3)  # 等待页面加载
    soup = BeautifulSoup(driver.page_source, 'lxml')

    # 查找新闻项
    pagebody_divs = soup.find_all('div', class_='pagebody')
    result_texts = []
    for div in pagebody_divs:
        # 找到所有class为pagebody-copy的div
        copy_divs = div.find_all('div', class_='pagebody-copy')
        
        # 提取文本并添加到结果列表
        for copy_div in copy_divs:
            result_texts.append(copy_div.get_text(strip=True))
    driver.back()
    time.sleep(3)  # 等待页面加载

    return result_texts
        

page_count = 0
while page_count < 228:
    # 获取页面的 HTML
    html = driver.page_source

    # 获取当前页的新闻并保存
    get_news_items(html)
    
    # 输出已完成当前页
    print(f"Page {page_count + 1} completed.")
    
    # 尝试找到“下一页”按钮
    try:
        # 使用 aria-label 或者 class 定位“下一页”按钮
        next_button = driver.find_element("xpath", "//a[@aria-label='next News articles']")  # 或者使用完整类名

        if next_button.is_displayed():
            next_button.click()  # 点击下一页
            page_count += 1  # 计数
            time.sleep(3)  # 等待页面加载
        else:
            print("no more page")
            break
    except Exception as e:
        print(f"no button to next page,warning: {e}")
        break


driver.quit()

with open('apple_news.csv', 'w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)  
    writer.writerow(['Title', 'Time', 'Content'])  
    writer.writerows(all_news_data)  

print("Data has been saved to apple_news.csv")
```
